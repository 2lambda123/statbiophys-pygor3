# Run IGoR using pygor3

Pygor works as an python interface for IGoR.

First import pygor3 package:

```python
import pygor3 as p3
```

Load a default model:

```python
species="human"
chain="tcr_beta"
mdl = p3.IgorModel.load_default(species, chain)
```

Use a model to align and assembly your sequences

```python
mdl.parms['v_choice']
```

```python
mdl.parms['d_gene']
```

```python
mdl.parms['j_choice']
```

## Evaluate Sequences

```python
# str_seq = str(df_gen_seqs['nt_sequence'].loc[0])[3*20:]
str_seq = "TTGAAATGTGAACAACATCTGGGTCATAACGCTATGTATTGGTACAAGCAAAGTGCTAAGAAGCCACTGGAGCTCATGTTTGTCTACAGTCTTGAAGAACGGGTTGAAAACAACAGTGTGCCAAGTCGCTTCTCACCTGAATGCCCCAACAGCTCTCACTTATCCCTTCACCTACACACCCTGCAGCCAGAAGACTCGGCCCTGTATCTCTGCGCCAGCAGCCGTAGGCGAGCGCGGCGGGGAGCTGTTTTTTGGAGAAGGCTCTAGGCTGACCGTACTGG"
str_seq
```

```python
help(p3.evaluate)
```

```python
df_scens = p3.evaluate(str_seq, mdl, N_scenarios=10)
df_scens
```

## Plotting an scenario

```python
ps_scenario = df_scens.iloc[1]
ps_scenario
mdl.plot_scenario(ps_scenario)
```

```python
df_scens, df_V_offsets = p3.evaluate(str_seq, mdl, N_scenarios=10, b_V_offset=True)
df_V_offsets
```

```python
ps_scenario = df_scens.iloc[0]
ps_scenario
seq_algn = (0, str_seq, -int(df_V_offsets.loc[0, ps_scenario['v_choice']]))
seq_algn
mdl.plot_scenario(ps_scenario, seq_aligned=seq_algn)
```

```python
ps_scenario = df_scens.iloc[3]
ps_scenario
seq_algn = (0, str_seq, -int(df_V_offsets.loc[0, ps_scenario['v_choice']]))
seq_algn
mdl.plot_scenario(ps_scenario, seq_aligned=seq_algn)
```

```python
p3.evaluate_pgen(str_seq, mdl)
```
